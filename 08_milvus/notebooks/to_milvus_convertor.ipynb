## Read and print data

```python
df
```

## Convert to milvus understandable standard

```python
def to_milvus(paper_id, parag_id, title, text):
    return {
        'title': f"{paper_id}#{parag_id}#{title}",
        'text': text
    }

converted_list = [to_milvus(*args) for args in zip(df['paper_id'], df['parag_id'],  df['title middle'],  df['text'])]
converted_df = pd.DataFrame(converted_list)
converted_df 
```

## Save converted data

```python
converted_df.to_csv('eval_parags_cleaned_mil.csv', index=False)
```
